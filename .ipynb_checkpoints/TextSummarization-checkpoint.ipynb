#### Preprocess text

```python
# Preprocess text
def text_preprocess(article0):
    #article0_processed=article0[article0.index('--')+2:]
    article0_processed = re.sub(r'\s+', ' ', article0)
    #article0_processed = re.sub(r'\s+\.\s+', ' ', article0_processed)
    #article0_processed = re.sub(r'\.\++', '', article0_processed)
    article0_processed = re.sub(r'[\:\-]+', '', article0_processed)
    #article0_processed=article0_processed.replace('\'', '').strip()
    article0_processed = re.sub(r'\s+\.\s+', ' ', article0_processed)
    return article0_processed
article0_processed=text_preprocess(article0)
article0_processed
```

```python
def text_in_lemma_form(article0_processed):
    nlp=spacy.load("en_core_web_md")
    text_lemma_list=[]
    doc=nlp(article0_processed)
    for token in doc:
        #print(token.lemma_)
        text_lemma_list.append(token.lemma_)
    text_lemma=" ".join(text_lemma_list)
    return text_lemma
text_in_lemma_form(article0_processed)
```

```python
# Removing special characters and digits
def text_preprocess_more(article0_processed):
    article0_processed_more=text_in_lemma_form(article0_processed)
    article0_processed_more = re.sub('[^a-zA-Z0-9]', ' ', article0_processed_more)
    article0_processed_more = re.sub(r'\s+', ' ', article0_processed_more)
    article0_processed_more=article0_processed_more.lower()
    return article0_processed_more
article0_processed_more=text_preprocess_more(article0_processed)
article0_processed_more
```

#### Cosine similarity in the sentence embeddings

```python
def cosine_similarity_matrix(sentences,embeddings):
    slen = len(sentences)
    sim_matrix = np.zeros((slen, slen))
    for sentence_a, embedding_a in zip(sentences, embeddings):
        for sentence_b, embedding_b in zip(sentences, embeddings):
            #print(a)
            #print(b)
            # compute cosine similarity
            cosine = np.dot(embedding_a,embedding_b)/(norm(embedding_a)*norm(embedding_b))
            #print("Cosine Similarity:", cosine)
            sentence_a_idx=sentences.index(sentence_a)
            sentence_b_idx=sentences.index(sentence_b)
            if sentence_a_idx != sentence_b_idx:
                sim_matrix[sentence_a_idx][sentence_b_idx] = cosine
    return sim_matrix
```

```python
sim_matrix=cosine_similarity_matrix(sentences,embeddings)
```

#### Page rank using sentence similarity matrix

```python
#https://towardsdatascience.com/text-summarization-in-python-with-jaro-winkler-and-pagerank-72d693da94e8
def page_ranked_sentence(sim_matrix,sentences):
    # create network
    #G = nx.from_numpy_matrix(sim_matrix)
    G = nx.from_numpy_array(sim_matrix)
    # calculate page rank scores
    #pr_sim = nx.pagerank(G)
    pr_sim = nx.pagerank_numpy(G)
    ranked_sentences = [
        #(sentences[sent], rank) for sent,rank in sorted(pr_sim.items(), key=lambda item: item[1], reverse = True)
        sent for sent,rank in sorted(pr_sim.items(), key=lambda item: item[1], reverse = True)
    ]
    return ranked_sentences,pr_sim
```

```python
ranked_sentences,sentence_scores=page_ranked_sentence(sim_matrix,sentences)
print(ranked_sentences)
print(sentence_scores)
```

```python
def combine_sentences(my_list,sentences):    
    summary=[]
    for i in sorted(my_list):
        s=sentences[i]
        #print(s)
        summary+=[s]
    summary=''.join(summary)
    return summary
```

```python
def cluster_kmeans(X,n_clusters):
    kmeans = KMeans(n_clusters, init = 'k-means++', random_state = 42)
    y_kmeans = kmeans.fit_predict(X)
    cluster_centers = kmeans.cluster_centers_
    return y_kmeans,cluster_centers
    """
    kmeans = KMeansClusterer(n_clusters, distance=nltk.cluster.util.cosine_distance,repeats=25,avoid_empty_clusters=True)
    y_kmeans = kmeans.cluster(X, assign_clusters=True)
    cluster_centers = [arr.tolist() for arr in kmeans.means()]
    return y_kmeans,cluster_centers
    """
```

```python
def fetch_top_n(X,y_kmeans,cluster_centers,nos):
    # Pick 1 sentence from each group
    my_list=[]
    for i in range(len(cluster_centers)):
        my_dict={}
        for j in range(len(y_kmeans)):
            if y_kmeans[j]==i:
                #my_dict[j] =  distance.euclidean(kmeans.cluster_centers_[i],[list(sentence_scores.values())[j]])
                my_dict[j] =  distance.euclidean(cluster_centers[i],X[j])
        #min_distance = min(my_dict.values())
        #my_list.append(min(my_dict, key=my_dict.get))
        my_list.extend([s for s,d in sorted(my_dict.items(), key=lambda item: item[1])][:nos])
    return my_list
```

### TF-IDF Model

#### Word Freq

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def getWordFreq(text):
    word_frequencies = {}
    for word in nltk.word_tokenize(text):
        if word not in stopwords:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1
    maximum_frequncy = max(word_frequencies.values())
    for word in word_frequencies.keys():
        word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
    return word_frequencies
```

```python
def getSentScore(word_frequencies,sentences):
    # Get sentence score by adding word frequency of all the words within the sentence
    sentence_scores = {}
    for i,sent in zip(range(len(sentences)),sentences):
        for word in nltk.word_tokenize(sent.lower()):
        #for word in nltk.word_tokenize(sent):
            if word in word_frequencies.keys():
                if i not in sentence_scores.keys():
                    sentence_scores[i] = word_frequencies[word]
                else:
                    sentence_scores[i] += word_frequencies[word]
        if i not in sentence_scores.keys():
            sentence_scores[i] = 0
    ranked_sentences = [
        #(sentences[sent], rank) for sent,rank in sorted(pr_sim.items(), key=lambda item: item[1], reverse = True)
        sent for sent,rank in sorted(sentence_scores.items(), key=lambda item: item[1], reverse = True)
    ]
    return ranked_sentences,sentence_scores
```

```python
def getSentenceFeatures(sentences,article_processed_more,highlights_processed):
    ## Dataframe Creation
    df = pd.DataFrame(sentences)
    df.columns=['sentence']
    #df['sentence_processed_more']=df.sentence.apply(text_preprocess_more)
    # Sentence position
    df['sentence_position']=df.index+1
    df['sentence_position']=df['sentence_position']/df['sentence_position'].max()
    # Sentence length
    df['sentence_length']=df.sentence.apply(lambda x: len(nlp(x)))
    # TF based sentence score
    word_frequencies=getWordFreq(article_processed_more)
    ranked_sentences,sentence_scores=getSentScore(word_frequencies,sentences)
    #print(sentence_scores)
    df['sentence_tf_avg']=sentence_scores.values()
    # POS ratio
    df['sentence_word_list']=df.sentence.apply(lambda x: [t for t in nlp(x)])
    df['sentence_pos_list']=df.sentence.apply(lambda x: [t.pos_ for t in nlp(x)])
    df['sentence_propn_count']=df['sentence_pos_list'].apply(lambda x: x.count('PROPN')/len(x))
    df['sentence_noun_count']=df['sentence_pos_list'].apply(lambda x: x.count('NOUN')/len(x))
    df['sentence_verb_count']=df['sentence_pos_list'].apply(lambda x: x.count('VERB')/len(x))
    df['sentence_adv_count']=df['sentence_pos_list'].apply(lambda x: x.count('ADV')/len(x))
    df['sentence_adj_count']=df['sentence_pos_list'].apply(lambda x: x.count('ADJ')/len(x))
    # Numbers in sentences
    df['sentence_num_count']=df['sentence_pos_list'].apply(lambda x: x.count('NUM')/len(x))
    #Stopwords in sentences
    df['sentence_stopword_count']=df['sentence'].apply(lambda x: sum([1 for t in nlp(x) if t.text in stopwords]))
    df['sentence_stopword_count']=df['sentence_stopword_count']/df['sentence_length']
    # Rogue-2 per sentence
    df['sentence_rougue2_f']=df['sentence'].apply(lambda x: rouge_metric(x, highlights_processed)[0]['rouge-2']['f'])
    return df
```

```python
def build_dataset_sample():
    dict_df={}
    i=0
    for article,highlights,sid in zip(dataset_sample['article'],dataset_sample['highlights'],dataset_sample['id']):
        print(i)
        i+=1
        article_processed=text_preprocess(article)
        article_processed_more=text_preprocess_more(article_processed)
        highlights_processed=text_preprocess(highlights)
        sentences=text_to_sentences(article_processed)
        df=getSentenceFeatures(sentences,article_processed_more,highlights_processed)
        df['sid']=sid
        #df_master=pd.concat([df_master,df])
        dict_df[sid]=df
        #break
    return dict_df
```

### SBERT+TopicModel+PageRank

```python
def topic_assignment(sentences,num_topics):
    ## Topic Model
    sentences_corpus=[]
    for s in sentences:
        s_processed_more=text_preprocess_more(s)
        words_corpus=[]
        for word in nltk.word_tokenize(s_processed_more):
            if word not in stopwords:
                words_corpus.append(word)
        sentences_corpus.append(words_corpus)
    # Create Dictionary
    id2word = corpora.Dictionary(sentences_corpus)
    # Term Document Frequency
    doc_term_matrix = [id2word.doc2bow(s) for s in sentences_corpus]
    # number of topics
    #num_topics = 2
    # Build LDA model
    lda_model = gensim.models.LdaMulticore(corpus=doc_term_matrix,id2word=id2word,num_topics=num_topics)
    doc_lda = lda_model[doc_term_matrix]
    labels=[]
    for i in doc_lda:
        dict_i=dict(i)
        #print(dict_i)
        l=[sent for sent,rank in sorted(dict_i.items(), key=lambda item: item[1], reverse = True)][0]
        labels.append(l)
    return labels
```
